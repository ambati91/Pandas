# 15. A Review of the .apply() Method on Single Columns

```python
bond = pd.read_csv('jamesbond.csv', index_col = 'Film')
bond.sort_index(inplace = True)
bond.head()
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + 'MILLIONS!'

bond['Box Office'].apply(convert_to_string_and_add_millions).head()
```

```python
bond['Bond Actor Salary'].apply(convert_to_string_and_add_millions).head()
```

```python
bond['Actor'].dtype
```

```python
bond['Bond Actor Salary'].dtype
```

```python
for i in bond.columns:
    if bond[i].dtype == 'float64':
        bond[i] = bond[i].apply(convert_to_string_and_add_millions)
```

```python
bond.head()
```

# 16. The .apply() Method with Row Values

```python
bond = pd.read_csv('jamesbond.csv',index_col = 'Film')
bond.sort_index(inplace = True)
bond.head()
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == 'Pierce Brosnan':
        return 'The Best'
    elif actor == 'Roger Moore' and budget > 40 :
        return 'Enjoyable'
    else:
        return 'I have No clue'

bond.apply(good_movie, axis = 'columns') # here we are looking from left to right thatswhy we have given axis = columns
#bond.head()
```
