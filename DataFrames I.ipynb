# Fill in Null values with the .fillna () Method 

```python
nba = pd.read_csv('nba.csv')
nba.head()
```

```python
nba.fillna(0).head() # it fills with zero wherever null values are there but this is not good practice, if you observe in 
                     # college column and index position 4, that also replace by zero. see below code to overcome this.
```

```python
nba['Salary'].fillna(0, inplace = True) # it replace with zero and modifies the original dataset if we use(inplace =True)
```

```python
nba.head() # if you observe the below output only salary column null values replaced by zero.
```

```python
nba['College'].fillna('No College', inplace = True) # it replaces 'No College' where null values in College columns
```

```python
nba.head()
```

# The .astype () Method

```python
nba = pd.read_csv('nba.csv').dropna(how = 'all') # drops if all values are null
nba['Salary'].fillna(0,inplace = True) # fill null values with zero and modifies or updates the existing dataframe
nba['College'].fillna('None',inplace = True) #fill null values with None and modifies or updates the existing dataframe
nba.head(6)
```

```python
nba.dtypes
```

```python
nba.info()
```

```python
nba['Salary']= nba['Salary'].astype('int') # converts salary column data types float to int and stores in Salary column
                                          # Note : .astype() doesn't have inplace parameter
```

```python
nba.head(3)
```

```python
nba['Age'] = nba['Age'].astype('int')
nba['Number'] = nba['Number'].astype('int')
nba.head(3)
# Note if you check nba.info() memory usage may decrease
```

```python
nba['Number'] = nba['Number'].astype('float') # converts int to float
nba.head(3)
```

```python
nba['Position'].nunique() # returns the number of unique values
# Note : for 5 unique values it was written 457 strings, to bring memory usage down we can use .astype("category") so performance
# may increase
```

```python
nba.info()
```

```python
nba['Position'] = nba['Position'].astype('category')
nba.info() # if you compare above and below memory usage it has come down
```

```python
nba.head()
```

# Sort a DataFrame with the .sort_values () Method, Part I

# Rank Values with the Rank () Method

```python
nba = pd.read_csv('nba.csv').dropna(how = 'all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
nba.head()
```

```python
nba['Salary'].rank().head() # if you look at index position 2 for that rank has given 6
```

```python
nba['Salar Rank'] = nba['Salary'].rank(ascending = False).astype('int')
```

```python
nba.head()
```

```python
nba.sort_values(by = 'Salar Rank', ascending = True).head()
```
