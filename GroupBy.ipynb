# 3. Retrieve A Group with the .get_group() Method

```python
fortune = pd.read_csv('fortune1000.csv', index_col = 'Rank')
fortune.head(3)
```

```python
sectors = fortune.groupby('Sector')
```

```python
sectors.get_group('Retailing').head(3) # it filters or displays only Retailing sectors
```

```python
# we can also extract above Retailing sectors as below

fortune[fortune['Sector'] == 'Retailing'].head(3)
```

```python
sectors.get_group('Energy').head(1) # extracting only energy sectors
```

```python
sectors.get_group('Technology').head(1) # extracting only technology related sectors
```

# 6. The .agg() Method

```python
fortune = pd.read_csv('fortune1000.csv', index_col = 'Rank')
fortune.head(3)
```

```python
sectors = fortune.groupby('Sector')
sectors
```

```python
sectors.sum().head()
```

```python
# if i want to perform for revenue column 'sum', for profit column 'mean', for employee column 'sum' we can do with the help of 
# agg method

sectors.agg({'Revenue': 'sum',
             'Profits': 'mean',
             'Employees' : 'sum'}).head()
```

```python
# if i want to calculate sum,mean,size for every column that can be done with agg method

sectors.agg(['sum','mean','size']).head()
```

# 7. Iterating through Groups

```python
fortune = pd.read_csv('fortune1000.csv', index_col = 'Rank')
fortune.head(3)
```

```python
sectors = fortune.groupby('Sector')
sectors
```

```python
df = pd.DataFrame(columns = fortune.columns)

for sector,data in sectors:
    highest_revenue_company_in_group = data.nlargest(1,'Revenue')
    df = df.append(highest_revenue_company_in_group)
df.head()
```

```python
for i,j in sectors: # here i = groupby first object and so on, and j = other than 'Sector' columns
    print(j)
    break
```
